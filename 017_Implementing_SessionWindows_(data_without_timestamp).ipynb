##Processing: session_windows.py

```python
import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions, StandardOptions
import os
from apache_beam import window
from apache_beam.transforms.trigger import AfterWatermark, AfterProcessingTime, AccumulationMode, AfterCount

# 📌 Ścieżka do pliku konta serwisowego Google Cloud (potrzebna do autoryzacji)
service_account_path = '/content/avon-prod-analytics.json'

# Wyświetlenie ścieżki do pliku (dla celów debugowania)
print("Service account file : ", service_account_path)

# Ustawienie zmiennej środowiskowej GOOGLE_APPLICATION_CREDENTIALS, aby Beam mógł się autoryzować w GCP
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = service_account_path

# 📌 Subskrypcja Pub/Sub, z której będą pobierane dane
input_subscription = 'projects/avon-prod-analytics/subscriptions/sub'

# 📌 Temat Pub/Sub, do którego będą zapisywane wyniki
output_topic = 'projects/avon-prod-analytics/topics/id-topics-2'

# 📌 Ustawienia dla Apache Beam
options = PipelineOptions()
options.view_as(StandardOptions).streaming = True  # Włączenie trybu przetwarzania strumieniowego

# 📌 Tworzenie głównej instancji potoku Apache Beam
p = beam.Pipeline(options=options)

# 📌 Funkcja konwertująca dane do formatu bajtowego (Pub/Sub wymaga bajtów)
def encode_byte_string(element):
    print (element)
    element = str(element)  # Zamiana wartości na string
    return element.encode('utf-8')  # Zakodowanie w formacie UTF-8 jako bajty

# 📌 Funkcja obliczająca zysk na podstawie ceny zakupu i sprzedaży
def calculateProfit(elements):
    buy_rate = elements[5]  # Pobranie ceny zakupu z listy
    sell_price = elements[6]  # Pobranie ceny sprzedaży
    products_count = int(elements[4])  # Pobranie liczby produktów (zamiana na int)

    # Obliczenie zysku: (cena sprzedaży - cena zakupu) * liczba produktów
    profit = (int(sell_price) - int(buy_rate)) * products_count

    elements.append(str(profit))  # Dodanie zysku jako ostatni element listy (string)
    return elements  # Zwrócenie zmodyfikowanej listy

pubsub_data= (
                p
                | 'Read from pub sub' >> beam.io.ReadFromPubSub(subscription=input_subscription)
                # STR_2,Mumbai,PR_265,Cosmetics,8,39,66/r/n
                | 'Remove extra chars' >> beam.Map(lambda data: (data.rstrip().lstrip()))
                # STR_2,Mumbai,PR_265,Cosmetics,8,39,66,
                | 'Split Row' >> beam.Map(lambda row: row.decode('utf-8').split(';'))
                # [STR_2,Mumbai,PR_265,Cosmetics,8,39,66,]
                | 'Filter By Country' >> beam.Filter(lambda elements : (elements[1] == "Mumbai" or elements[1] == "Bangalore"))
                | 'Create Profit Column' >> beam.Map(calculateProfit)
                # [STR_2,Mumbai,PR_265,Cosmetics,8,39,66,27]
                | 'Form Key Value pair' >> beam.Map(lambda elements : (elements[0], int(elements[7])))
                # STR_2 27
                # | 'Window' >> beam.WindowInto(window.FixedWindows(20)) # FixedWindows
                # | 'Window' >> beam.WindowInto(window.SlidingWindows(30,10)) # SlidingWindows
                | 'Window' >> beam.WindowInto(window.Sessions(25)) # Sessions
                | 'Sum values' >> beam.CombinePerKey(sum)
                | 'Encode to byte string' >> beam.Map(encode_byte_string)
                | 'Write to pus sub' >> beam.io.WriteToPubSub(output_topic)
	             )

result = p.run()
result.wait_until_finish()
```

##subscribe.py

```python
from google.cloud import pubsub_v1
import time
import os

if __name__ == "__main__":

    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = '/content/avon-prod-analytics.json'

    subscription_path = 'projects/avon-prod-analytics/subscriptions/sub-2'

    subscriber = pubsub_v1.SubscriberClient()

    def callback(message):
        print(('Received message: {}'.format(message)))
        message.ack()

    subscriber.subscribe(subscription_path, callback=callback)

    while True:
        time.sleep(60)
```
