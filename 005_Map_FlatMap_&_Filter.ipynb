##output_new_final_2

```python
def SplitRow(element):
    return element.split(',')

p1 = beam.Pipeline()

attendance_count = (

   p1
    |beam.io.ReadFromText('/content/005_dept_data.txt')
    |beam.Map(SplitRow)
    |beam.io.WriteToText('data/output_new_final_2)')

)

p1.run()
```

##output_new_final_3

```python
def SplitRow(element):
    return element.split(',')

def filtering(record):
  return record[3] == 'Accounts'

p1 = beam.Pipeline()

attendance_count = (

   p1
    |beam.io.ReadFromText('/content/005_dept_data.txt')
    |beam.Map(SplitRow)
    |beam.Filter(filtering)
    |beam.io.WriteToText('data/output_new_final_3)')

)

p1.run()
```

```python
!{("head -n 20 '/content/data/output_new_final_3)-00000-of-00001'")}
```

##output_new_final_4

```python
def SplitRow(element):
    return element.split(',')

def filtering(record):
  return record[3] == 'Accounts'

p1 = beam.Pipeline()

attendance_count = (

   p1
    |beam.io.ReadFromText('/content/005_dept_data.txt')
    |beam.Map(SplitRow)
    |beam.Filter(filtering)
    |beam.Map(lambda record: (record[1], 1))
    |beam.io.WriteToText('data/output_new_final_4)')

)

p1.run()
```

##output_new_final

```python
def SplitRow(element):
    return element.split(',')

def filtering(record):
  return record[3] == 'Accounts'


p1 = beam.Pipeline()

attendance_count = (

   p1
    |'Odczyt z pliku' >> beam.io.ReadFromText('/content/005_dept_data.txt')
    |'Transformacja MAP' >> beam.Map(SplitRow)
   #| beam.Map(lambda record: record.split(','))

    |beam.Filter(filtering)
  # |beam.Filter(lambda record: record[3] == 'Accounts')

    |beam.Map(lambda record: (record[1], 1))
    |beam.CombinePerKey(sum)

    |beam.io.WriteToText('data/output_new_final')

)

p1.run()
```
